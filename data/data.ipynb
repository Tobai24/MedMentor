### Download dataset from hugging face

```python
from huggingface_hub import login

login(token="")
```

```python
from datasets import Features, Image, Value , load_dataset

# Define the expected features
features = Features({
    "image": Image(),  # Image field
    "report": Value("string"),  # Radiology report
    "view_position": Value("string"),  # View position (e.g., PA, AP, Lateral)
})

# Reload the dataset with the correct features
dataset = load_dataset("MK-runner/Multi-view-CXR", features=features)
```
